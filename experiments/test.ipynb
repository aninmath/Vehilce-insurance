```python
def anierrorfunc (error: Exception, error_detail: sys):
    return(f"error happened in the file details: {str(error)}")

class aniexception(Exception):
    def __init__(self,error: Exception, error_detail: sys):
        super().__init__(error)
        self.errror= anierrorfunc (error, error_detail )

    def __str__(self):
        return (f'this is cause im modifying string with {self.errror}')
```

```python
import sys

try:       
    a = 1+'Z'
except Exception as e:
      raise aniexception(e,sys) from e
```

```python
class aniclass():
    def __init__(self, *arg):
        self.varibale = arg

    def __str__(self):
        return (f'this class has comprised of {[x for x in self.varibale]}')
```

```python
from dataclasses import dataclass

@dataclass

class Peroson:
    name: str = 'Ani'
    age : int = 23
```

```python
name: Peroson = Peroson ('Oli')
```
